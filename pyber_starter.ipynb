```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combine_data = pd.merge(city_data, ride_data, how="left", on=["city","city"])

# Display the data table for preview
print(combine_data)
```

```python
city_df = combine_data[["city", "date", "fare", "ride_id", "driver_count", "type"]]
city_df.head()

city_df.groupby("city")
city_df.head()
```

```python
urban = combine_data[(combine_data["type"] == "Urban")]
rural = combine_data[(combine_data["type"] == "Rural")]
suburban = combine_data[(combine_data["type"] == "Suburban")]
print(urban)
print(rural)
print(suburban)
```

```python
avg_urban = urban.groupby("city")["fare"].mean()
avg_rural = rural.groupby("city")["fare"].mean()
avg_suburban = suburban.groupby("city")["fare"].mean()
print(avg_urban)
print(avg_rural)
print(avg_suburban)
```

```python
ride_count_urban = urban.groupby(["city"]).count()["ride_id"]
ride_count_rural = rural.groupby(["city"]).count()["ride_id"]
ride_count_suburban = suburban.groupby(["city"]).count()["ride_id"]
print(ride_count_urban)
print(ride_count_rural)
print(ride_count_suburban)
```

```python
driver_count_urban = urban.groupby("city")["driver_count"].mean()
driver_count_rural = rural.groupby("city")["driver_count"].mean()
driver_count_suburban = suburban.groupby("city")["driver_count"].mean()
print(driver_count_urban)
print(driver_count_rural)
print(driver_count_suburban)
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
x_axis = [ride_count_urban, ride_count_rural, ride_count_suburban]
y_axis = [avg_urban, avg_rural, avg_suburban]

# Build the scatter plots for each city types
plot_urban = plt.scatter(ride_count_urban, avg_urban, linewidths=1, marker="o", facecolor="gold", edgecolor="black", s=10*driver_count_urban, alpha=0.8, label="Urban")
plot_rural = plt.scatter(ride_count_rural, avg_rural, linewidths=1, marker="o", facecolor="skyblue", edgecolor="black", s=7*driver_count_rural, alpha=0.8, label="Rural")
plot_suburban = plt.scatter(ride_count_suburban, avg_suburban, linewidths=1, marker="o", facecolor="coral", edgecolor="black", s=8*driver_count_suburban, alpha=0.8, label="Suburban")

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

plt.grid(True)

# Create a legend
lgnd=plt.legend(["Urban", "Rural", "Suburban"], title="City Types", loc='upper right')
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(42,35,"Note:\n Circle size correlates with driver count per city") 

# Save Figure
```

```python
# Show plot
plt.show()
```
